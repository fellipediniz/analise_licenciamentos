Gráfico que está realizando a comparação da quantidade de carros por tipo de combustível, onde podemos concluir que nos anos de 2022 e 2023 os combustíveis do tipo Flex Fuel e Diesel, seguiram sendo os mais consumidos. Além de conseguir visualizar, o aumento no consumo de carros elétricos, tendo uma pequena crescente, porém significativa.

```python
#Transformando a coluna de datas em dateTime
df_veiculos_combustivel['data'] = pd.to_datetime(df_veiculos_combustivel['data'], format='%d-%m-%y')
df_veiculos_combustivel['Ano'] = df_veiculos_combustivel['data'].dt.year
df_veiculos_combustivel['Mes'] = df_veiculos_combustivel['data'].dt.strftime('%b')
#excluindo meses que não existem nos dois anos
df = df_veiculos_combustivel[~df_veiculos_combustivel['Mes'].isin(['Nov', 'Dec'])]
#agrupando por mês e ano
df_agrupado_mes = df.groupby(['Ano', 'Mes']).mean().reset_index()

#configurações do plot
fig, ax = plt.subplots(figsize=(15,8))
plt.title('Número de licenciamentos por mês em função do tipo de combustível')
plt.xlabel('Ano')
plt.ylabel('Quantidade de carros')
plt.xticks(rotation=0)

#loop para obtenção de valores mensais de cada ano por tipo de combustível
for coluna in ['Diesel', 'Gasolina', 'Flex Fuel', 'Elétrico', 'Híbrido']:
    for ano in df_agrupado_mes['Ano'].unique():
        df_ano = df_agrupado_mes[(df_agrupado_mes['Ano'] == ano)]
        if ano == 2022:
            ax.plot(df_ano['Mes'], df_ano[coluna], marker='o', linestyle='dotted', label=f'{coluna} {ano}')
        else:
            ax.plot(df_ano['Mes'], df_ano[coluna], marker='o', label=f'{coluna} {ano}')


ax.legend()
plt.show()
```

Esse gráfico mostra a relação do número de vendas por mês em cada ano, através desse gráfico pode-se observar um comportamento semelhante acerca dos números de licenciamentos. Dando destaque aos meses de menor valor, janeiro e fevereiro, e aos com maior número, julho e outubro.

```python
df_agrupado_mes.dtypes
```

```python
#criando um df exclusivo para tipo 'leve e comerciais'
df_leves = df_veiculos_combustivel[df_veiculos_combustivel['tipo'] == 'leves e comerciais']

#separando o df por por combustível
df_leves_colunas = df_leves.melt(id_vars=['ano', 'tipo'], value_vars=['Diesel', 'Gasolina', 'Flex Fuel', 'Elétrico', 'Híbrido'], var_name='combustivel', value_name='contagem')

#obtendo valores percentuais e reais
df_leves_colunas['var_percentual'] = df_leves_colunas.groupby(['combustivel'])['contagem'].pct_change()
df_grouped = df_leves_colunas.groupby(['ano', 'combustivel']).sum().reset_index()

#configurações do plot
fig, ax = plt.subplots(figsize=(15, 8))
bar_width = 0.18
plt.xlabel('Ano', fontweight='bold')
plt.ylabel('Quantidade')
plt.title('Quantidade de automóveis leves e comerciais licenciados por ano em função do combustiel utilizado')

#loop para os tipos de combustível e colocação dos valores reais e porcentagem
for i, combustivel in enumerate(df_grouped['combustivel'].unique()):
    df_combustivel = df_grouped[df_grouped['combustivel'] == combustivel]

    bars = ax.bar(np.arange(len(df_combustivel['ano'])) + i * bar_width, df_combustivel['contagem'], width=bar_width, label=combustivel)
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.0f}', ha='center', va='bottom')
        ax.text(bar.get_x() + bar.get_width()/2, yval/2, f'{yval/df_combustivel["contagem"].sum()*100:.2f}%', ha='center', va='center', color='black')

ax.legend()


ax.set_xticks(np.arange(len(df_combustivel['ano'])) + bar_width / 2)
ax.set_xticklabels(df_combustivel['ano'])


plt.show()
```

Gráfico que está realizando a comparação da quantidade de licenciamentos por categoria (leves e comerciais) e por tipo de combustível, onde podemos concluir que nos anos de 2022 e 2023 os combustíveis do tipo Flex Fuel e Diesel, seguiram sendo os mais consumidos. Além de conseguir visualizar, o aumento no consumo de carros elétricos, tendo uma pequena crescente, porém significativa.

```python
#criando um df exclusivo para tipo 'caminhoes e onibus'
df_pesados = df_veiculos_combustivel[df_veiculos_combustivel['tipo'] == 'caminhoes e onibus']

#separando o df por por combustível
df_pesados_colunas = df_pesados.melt(id_vars=['ano', 'tipo'], value_vars=['Diesel', 'Gasolina', 'Flex Fuel', 'Elétrico', 'Híbrido'], var_name='combustivel', value_name='contagem')

#obtendo valores percentuais e reais
df_pesados_colunas['var_percentual'] = df_pesados_colunas.groupby(['combustivel'])['contagem'].pct_change()
df_pesados_colunas = df_pesados_colunas.fillna(0)
df_pesados_colunas['var_percentual'] = df_pesados_colunas['var_percentual'].fillna(0)

df_grouped = df_pesados_colunas.groupby(['ano', 'combustivel']).sum().reset_index()

#configurações do plot
fig, ax = plt.subplots(figsize=(15, 8))
bar_width = 0.18
plt.xlabel('Ano', fontweight='bold')
plt.ylabel('Quantidade')
plt.title('Quantidade de automóveis caminhões e ônibus licenciados por ano em função do combustiel utilizado')

# Loop sobre os tipos de combustível
for i, combustivel in enumerate(df_grouped['combustivel'].unique()):
    df_combustivel = df_grouped[df_grouped['combustivel'] == combustivel]

    bars = ax.bar(np.arange(len(df_combustivel['ano'])) + i * bar_width, df_combustivel['contagem'], width=bar_width, label=combustivel)

    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.0f}', ha='center', va='bottom')
        ax.text(bar.get_x() + bar.get_width()/2, yval/2, f'{yval/df_combustivel["contagem"].sum()*100:.2f}%', ha='center', va='center', color='black')

ax.legend()
ax.set_xticks(np.arange(len(df_combustivel['ano'])) + bar_width / 2)
ax.set_xticklabels(df_combustivel['ano'])

plt.show()
```

Gráfico que está realizando a comparação da quantidade de licenciamentos por categoria (caminhões e ônibus) e por tipo de combustível, onde podemos concluir que nos anos de 2022 e 2023 que o combustível Diesel, representa quase todos os licenciamentos referentes a essa categoria. Também pode-se observar que um queda de aproximadamente 60% nos licenciamentos desses veículos de fonte energética elétrica.

```python
#Tratando os dados do df
df_veiculos_fabricacao.rename(columns={'mês': 'mes'}, inplace=True)
df_veiculos_fabricacao = df_veiculos_fabricacao.drop('tipo', axis=1)
#agrupando dados por mês, ano, categoria e fabricação
df_final_fabricacao = df_veiculos_fabricacao.groupby(['mes', 'ano', 'Categoria', 'Fabricação']).sum().reset_index()
#removendo os mêses que não estão contidos nos 2 anos
df_veiculos_fabricacao = df_final_fabricacao[~((df_final_fabricacao['mes'] == 'Nov') | (df_final_fabricacao['mes'] == 'Dez'))]
```

```python
df_final_fabricacao
```

```python
#modificando o tipo da coluna
df_final_fabricacao['ano'] = df_final_fabricacao['ano'].astype(str)

#reclassificando a coluna de categoria
df_final_fabricacao['Categoria'] = df_final_fabricacao['Categoria'].replace({'Ônibus': 'caminhoes e onibus', 'Caminhões': 'caminhoes e onibus'})
df_final_fabricacao['Categoria'] = df_final_fabricacao['Categoria'].replace({'Veículos leves': 'leves e comerciais'})
#agrupando novos dados
df_final_fabricacao = df_final_fabricacao.groupby(['mes', 'ano', 'Categoria', 'Fabricação']).sum().reset_index()
```

```python
df_final_fabricacao
```

```python
# Filtrando os dados para as categorias desejadas
df_leves_comerciais = df_final_fabricacao[df_final_fabricacao['Categoria'] == 'leves e comerciais']
df_caminhao_onibus = df_final_fabricacao[df_final_fabricacao['Categoria'] == 'caminhoes e onibus']

# Agrupando por ano e fabricação e somar as quantidades
df_leves_comerciais = df_leves_comerciais.groupby(['ano', 'Fabricação']).sum().reset_index()
df_caminhao_onibus = df_caminhao_onibus.groupby(['ano', 'Fabricação']).sum().reset_index()

#configurando as barras
barWidth = 0.15
r1 = np.arange(len(df_leves_comerciais['ano'].unique()))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]


plt.figure(figsize=(15, 8))

# Plotar os dados converificação de dados vazios
if not df_leves_comerciais[df_leves_comerciais['Fabricação'] == 'Nacional'].empty:
    bars1 = plt.bar(r1, df_leves_comerciais[df_leves_comerciais['Fabricação'] == 'Nacional']['Quantidade'], width=barWidth, label='Leves e Comerciais - Nacional')
if not df_leves_comerciais[df_leves_comerciais['Fabricação'] == 'Internacional'].empty:
    bars2 = plt.bar(r2, df_leves_comerciais[df_leves_comerciais['Fabricação'] == 'Internacional']['Quantidade'], width=barWidth, label='Leves e Comerciais - Internacional')
if not df_caminhao_onibus[df_caminhao_onibus['Fabricação'] == 'Nacional'].empty:
    bars3 = plt.bar(r3, df_caminhao_onibus[df_caminhao_onibus['Fabricação'] == 'Nacional']['Quantidade'], width=barWidth, label='Ônibus e Caminhão - Nacional')
if not df_caminhao_onibus[df_caminhao_onibus['Fabricação'] == 'Internacional'].empty:
    bars4 = plt.bar(r4, df_caminhao_onibus[df_caminhao_onibus['Fabricação'] == 'Internacional']['Quantidade'], width=barWidth, label='Ônibus e Caminhão - Internacional')

#função de loop  para posicionar valores nas barras
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height,
                 '%d' % int(height), ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)
add_labels(bars4)

plt.xlabel('Ano')
plt.ylabel('Quantidade Total Anual')
plt.title('Quantidade de Licenciamentos por Categoria e Fabricação')
plt.xticks([r + barWidth for r in range(len(df_leves_comerciais['ano'].unique()))], df_leves_comerciais['ano'].unique())
plt.legend()
plt.show()
```

Nesse gráfico é mostrado a quantidade de veículos licenciados referentes a categoria e local de fabricação, fica expresso que os veículos mais licenciados são os que são produzidos no Brasil. Quando tratamos dos contidos na categoria dos "leves e comerciais" observa-se que ocorreu um avanço nos licenciamentos tanto dos nacionais quanto dos importados. Comportamento diferente da categoria dos caminhões e ônibus que passou por uma redução no número de nacionais e aumento no número de importados.

```python
# Filtrando os dados para a categoria desejada
df_leves_comerciais = df_final_fabricacao[df_final_fabricacao['Categoria'] == 'leves e comerciais']

# Agrupando por ano, mês e fabricação e somar as Quantidades
df_leves_comerciais = df_leves_comerciais.groupby(['ano', 'mes', 'Fabricação']).sum().reset_index()

# Plotar os dados
plt.figure(figsize=(10, 6))

for ano in df_leves_comerciais['ano'].unique():
    for fabricacao in df_leves_comerciais['Fabricação'].unique():
        df_ano_fabricacao = df_leves_comerciais[(df_leves_comerciais['ano'] == ano) & (df_leves_comerciais['Fabricação'] == fabricacao)]
        if not df_ano_fabricacao.empty:
            plt.plot(df_ano_fabricacao['mes'], df_ano_fabricacao['Quantidade'], marker='o', linestyle='-' if ano != 2022 else '--', label=f'{ano} - {fabricacao}')

plt.xlabel('Mês')
plt.ylabel('Quantidade Total Anual')
plt.title('Quantidade de Licenciamentos por Mês e Fabricação de Veículos Leves')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Nesse gráfico é possível observar as variações do número de licenciamentos mensais de automóveis leves e comerciais nos anos de 2022 e 2023. Vale ressaltar a similaridade de comportamento entre os anos, mesmo existindo pontos de discrepância o número de licenciamentos tende a seguir picos em meses específicos nos anos citados.   

```python
# Filtrar os dados para a categoria desejada
df_caminhao_onibus = df_final_fabricacao[df_final_fabricacao['Categoria'] == 'caminhoes e onibus']

# Agrupando por ano, mês e fabricação e somar as Quantidades
df_caminhao_onibus = df_caminhao_onibus.groupby(['ano', 'mes', 'Fabricação']).sum().reset_index()

# Plotar os dados
plt.figure(figsize=(10, 6))

for ano in df_caminhao_onibus['ano'].unique():
    for fabricacao in df_caminhao_onibus['Fabricação'].unique():
        df_ano_fabricacao = df_caminhao_onibus[(df_caminhao_onibus['ano'] == ano) & (df_caminhao_onibus['Fabricação'] == fabricacao)]
        if not df_ano_fabricacao.empty:
            plt.plot(df_ano_fabricacao['mes'], df_ano_fabricacao['Quantidade'], marker='o', linestyle='-' if ano != 2022 else '--', label=f'{ano} - {fabricacao}')

plt.xlabel('Mês')
plt.ylabel('Quantidade Total Anual')
plt.title('Quantidade de Licenciamentos por Mês e Fabricação de Caminhões e Onibus')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Um pouco discrepante do gráfico acima, o licenciamento de caminhões e ônibus, mostra um comportamento discrepante para os anos de 2022 e 2023 quando tratamos de automóveis nacionais. Por outro o comportamento do licenciamento dos veículos internacionais tende a seguir parelhos.

Com base nos dados obtidos, pode-se concluir que o licenciamento referente aos anos de 2022 e 2023 passou por uma mudança. Embora ainda expressivo, o número de licenciamentos de automóveis de fontes energéticas não renováveis, reduziu de forma eloquente.
Também pode-se destacar o aumento dos licenciamentos de automóveis de fontes energéticas renováveis, fato esse que não se deve relacionar exclusivamente a mudança no comportamento do consumidor, pois políticas públicas viabilizaram a compra desse tipo de automóveis.
